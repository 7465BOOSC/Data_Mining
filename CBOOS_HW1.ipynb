```python
#Connor Boos
#Homework 1


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import statsmodels.api as sm

```

```python
df = pd.read_csv("iris.csv", header = None)
#Change the headers to be the attributes
df.columns = ['Sepal_Length', 'Sepal_Width', 'Petal_Length', 'Petal_Width', 'Class']
```

Beginning the process of visualizing data

```python
#Part 1: Computing the Standard Deviation of each attribute
df.describe()
```

This shows that the standard deviation for each attribute is as follows:
Sepal Length: .0828066
Sepal Width: .0433594
Petal Length: 1.64420
Petal Width: .763161

```python
#Compute Boxplot for Sepal Length

sns.boxplot(y = df.Sepal_Length)
plt.show()
```

The boxplot shows that the sepal length has a max of ~7.9 and a minimum of ~4.3 the median value is ~5.7 and q1 is ~5.1 and q3 is ~6.4

```python
#Compute Boxplot for Sepal Width
sns.boxplot(y = df.Sepal_Width)
plt.show()
```

The boxplot shows that the sepal width has a max of ~4.0 and a minimum of ~2.2. There are several outliers at ~ 2, 4.1, 4.2, 4.4 the median value is ~3.0 and q1 is ~2.7 and q3 is ~3.4

```python
#Compute Boxplot for Petal Width
sns.boxplot(y = df.Petal_Width)
plt.show()
```

The boxplot shows that the petal width has a max of ~2.5 and a minimum of ~.1 the median value is ~1.3 and q1 is ~.3 and q3 is ~1.7

The box plots where the median is not centrally located in between Q1 and Q3 have the largest variance

```python

#Compute Boxplot for Petal Length
sns.boxplot(y = df.Petal_Length)
plt.show()
```

The boxplot shows that the sepal width has a max of 6.9 and a minimum of 1 the median value is ~4.4 and q1 is ~1.7 and q3 is ~5.0

```python
#Get the variance of each column
df.var()
```

The Largest variance belongs to Petal Length.
There are outliers in the Sepal Width data. At values [2, 4.1, 4.2, 4.4]

```python
#Part 2: Compute Histogram for each Attribute

#Histogram for Sepal Length
sns.distplot(df.Sepal_Length)
plt.show()
```

```python
#Histogram for Sepal Width
sns.distplot(df.Sepal_Width)
plt.show()
```

```python
#Histogram for Petal Length
sns.distplot(df.Petal_Length)
plt.show()
```

```python
#Histogram for Petal Width
sns.distplot(df.Petal_Width)
plt.show()
```

Sepal Width shows the distribution closest to a normal bellcurve. And sepal length is close to normal but not exactly.
Petal Length and Petal Width are bimodal and do not show a normal bellcurve

```python
#Part 3: 
#QQ Plot for Sepal Width
stats.probplot(df.Sepal_Width, dist="norm", plot=plt)
plt.show()
```

```python
# QQ Plot for Sepal Length
stats.probplot(df.Sepal_Length, dist="norm", plot=plt)
plt.show()
```

When all the points fall on the line it show that the data has a normal distribution. These QQ plots show that a majority of the points fall on the line so the distributions are approximately normal. This follows the trend of the histogram plots and the box plots for both of these parts of the data.

```python
# Part 4: Scatter Plots for the Attributes
# Sepal Length & Sepal Width
plt.scatter(df.Sepal_Length, df.Sepal_Width)
plt.title("Sepal Width vs. Sepal Length")
plt.xlabel("Sepal Width")
plt.ylabel("Sepal Length")
plt.show()
```

The scatter shows a weak neagtive correlation between the variables.

The scatter shows that there is a moderately strong positive correlation between the variables.

The scatter plot shows a weak negative correlation between the variables.
